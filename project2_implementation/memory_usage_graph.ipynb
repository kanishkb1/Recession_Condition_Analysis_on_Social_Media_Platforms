```python
#!/usr/bin/env python3
import os
import sys
import json
import time
import time
import base64
import requests
import pickle
import configparser
import numpy as np
from dateutil.parser import parse
from datetime import datetime, date
from matplotlib import pyplot as plt, dates as mdates

def plot_data_size(size_bytes, data_name):

    days = np.arange(1,22)
    days_bytes = []
    for b_size in size_bytes:
        days_bytes.append((days*b_size)/1000000)
    plt.figure(figsize=(8, 8))
    plt.plot(days, days_bytes[2], ".--", label=data_name[2])
    plt.plot(days, days_bytes[1], "o--", label=data_name[1])
    plt.plot(days, days_bytes[0], "x--", label=data_name[0])
    plt.grid(True)
    plt.yscale("log", basey=10)
    plt.title("Memory Usage: Twitter, Reddit, NYTimes")
    plt.xlabel("Number of Days (Date Range: 10/31/2022 to 11/21/2022)")
    plt.ylabel("Dataset Size (in Megabytes)")
    plt.legend(loc="best")
    plt.tight_layout()
    # plt.show()
    plt.savefig("size_graph.png")

    return 0

def main():
    size_bytes = [40580000, 62590000, 49580000]
    data_name = ["Twitter", "Reddit", "NYTimes"]
    plot_data_size(size_bytes, data_name)
    return 0

if __name__ == '__main__':
    main()
```
